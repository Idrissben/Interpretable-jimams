## Import data

```python
raw_data = pd.read_csv("data/data_clean.csv")
df_one_hot = data_utils.preprocess_data(raw_data)
```

## First EDA

```python
df_one_hot.info()
```

```python
# Count the occurrences of True and False
y_counts = df_one_hot['y'].value_counts()

# Create a bar plot to show the distribution
plt.figure(figsize=(6, 4))
plt.bar(y_counts.index, y_counts)
plt.xticks([True, False], ['Release', 'Denial'])
plt.title('Distribution of Parole Outcomes')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.show()
```

```python
# Count the occurrences of True and False
y_counts = df_one_hot['sex_MALE'].value_counts()

# Create a bar plot to show the distribution
plt.figure(figsize=(6, 4))
plt.bar(y_counts.index, y_counts)
plt.xticks([0, 1], ['Female', 'Male'])
plt.title('Distribution of Gender')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.show()
```

```python
# Sum the occurrences of each race/ethnicity category
race_counts = df_one_hot[['race / ethnicity_ASIAN/PACIFIC',
                  'race / ethnicity_BLACK',
                  'race / ethnicity_HISPANIC',
                  'race / ethnicity_OTHER',
                  'race / ethnicity_UNKNOWN',
                  'race / ethnicity_WHITE']].sum()
# Sort the counts in descending order
race_counts = race_counts.sort_values(ascending=False)

# Create a bar plot
plt.figure(figsize=(10, 6))
race_counts.plot(kind='bar')
plt.title('Distribution of Race/Ethnicity Categories')
plt.xlabel('Race/Ethnicity Categories')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
